```python
def calc_position(arr):
    store = {}
    for i in range(len(arr)):
        data = arr[i].split(" ")
        if data[0] in store:
            store[data[0]] += int(data[1])
        else:
            store[data[0]] = int(data[1])
    res = {}
    res["h"] = store["forward"]
    res["d"] = store["down"] - store["up"]
    return res["h"] * res["d"]
    
```

```python
def calc_position_aim(arr):
    store = {
        "H" : 0,
        "D" : 0,
        "A" : 0
    }
    for i in range(len(arr)):
        data = arr[i].split(" ")
        if data[0] == "forward":
            # update horizontal
            store["H"] += int(data[1])
                
            # update depth
            if store["A"] != 0:
                store["D"] += store["A"] * int(data[1])
        if data[0] == "down":
            store["A"] += int(data[1])
        if data[0] == "up":
            store["A"] -= int(data[1])
    return store["H"] * store["D"]
                
```
