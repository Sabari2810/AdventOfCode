```python
def get_least_or_most_common_bit(bits,n,isLeast = False):
    store = {"0" : 0,"1" : 0}
    val = ""
    if n < 0:
        return val
    val = "";
    for i in range(len(bits)):
        store[bits[i][n]] += 1
    if isLeast:
        if store["0"] < store["1"]:
            val = "0"
        else:
            val = "1"
    else: 
        if store["0"] > store["1"]:
            val = "0"
        else:
            val = "1"
    return val + get_least_or_most_common_bit(bits,n-1,isLeast)
```

```python
def get_gamma_rate(bits):
    return int(get_least_or_most_common_bit(bits,len(bits[0]) - 1)[::-1],2)
```

```python
def get_epsilon_rate(bits):
    return int(get_least_or_most_common_bit(bits,len(bits[0]) - 1,True)[::-1],2)
```

```python
def power_consumption(bits):
    return get_gamma_rate(bits) * get_epsilon_rate(bits)
```

```python
def get_rating_bit(bits,n,isCo2 = False):
    store = {"0" : 0,"1" : 0}
    new_bits = []
    if len(bits) == 1:
        return bits
    for i in range(len(bits)):
        store[bits[i][n]] += 1
    for j in range(len(bits)):
        if isCo2:
            if store["0"] <= store["1"]:
                if bits[j][n] == "0":
                    new_bits.append(bits[j])
            else:
                if bits[j][n] == "1":
                    new_bits.append(bits[j])
        else:
            if store["0"] > store["1"]:
                if bits[j][n] == "0":
                    new_bits.append(bits[j])
            else:
                if bits[j][n] == "1":
                    new_bits.append(bits[j])
    return get_rating_bit(new_bits,n+1,isCo2)
```

```python
def get_oxygen_generator_rating(bits):
    return int(get_rating_bit(bits,0)[0],2)
```

```python
def get_co2_scrubber_rating(bits):
    return int(get_rating_bit(bits,0,True)[0],2)
```
