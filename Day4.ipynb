```python
def has_complete_Column(grid,placeholder):
        n = 4
        while n >= 0:
            d_map = {}
            for i in range(5):
                if grid[i][n] in d_map:
                    d_map[grid[i][n]] += 1
                else:
                    d_map[grid[i][n]] = 1
            if placeholder in d_map:        
                if d_map[placeholder] == 5:
                    return True
            n -= 1
        return False
```

```python
def has_complete_Row(grid,placeholder):
    for i in range(5):
        d_map = {}
        for n in range(5):
            if grid[i][n] in d_map:
                d_map[grid[i][n]] += 1
            else:
                d_map[grid[i][n]] = 1
        if placeholder in d_map:        
            if d_map[placeholder] == 5:
                return True
    return False
```

```python
def bingo(grids,n,placeholder):
    for i in range(len(grids)):
        for j in range(5):
            for k in range(5):
                if grids[i][j][k] == n:
                    grids[i][j][k] = placeholder
        if has_complete_Column(grids[i],placeholder):
            return i
        if has_complete_Row(grids[i],placeholder):
            return i
    return False
```

```python
def get_grid_sum(grid,placeholder):
    sum = 0
    for i in range(len(grid)):
        for j in range(len(grid[i])):
            if grid[i][j] != placeholder:
                sum += grid[i][j]
    return sum
```

```python
def get_first_board(nums,grids,placeholder):
    for num in nums:
        res = bingo(grids,num,placeholder)
        if res != False:
            return get_grid_sum(grids[res],placeholder) * num
```
